```python
import json
import os

# File to store data
DATA_FILE = "students.json"

# Load data from file
def load_data():
    if os.path.exists(DATA_FILE):
        with open(DATA_FILE, "r") as f:
            try:
                return json.load(f)
            except json.JSONDecodeError:
                return {}
    return {}

# Save data to file
def save_data(data):
    with open(DATA_FILE, "w") as f:
        json.dump(data, f, indent=4)

# Initialize dictionary
student_grades = load_data()

# Add student
def add_student(name, grade):
    student_grades[name] = grade
    save_data(student_grades)
    print(f" Added {name} with grade {grade}")

# Update student
def update_student(name, grade):
    if name in student_grades:
        student_grades[name] = grade
        save_data(student_grades)
        print(f" {name}'s grade updated to {grade}")
    else:
        print(f" Student '{name}' not found")

# Delete student
def delete_student(name):
    if name in student_grades:
        del student_grades[name]
        save_data(student_grades)
        print(f" {name} deleted successfully")
    else:
        print(f" Student '{name}' not found")

# Display all students
def display_all_students():
    if student_grades:
        print("\n Student List:")
        for name, grade in student_grades.items():
            print(f" - {name}: {grade}")
    else:
        print("No students found")

# Input validation for grades
def get_grade_input():
    while True:
        try:
            grade = int(input("Enter the grade (0â€“100): "))
            if 0 <= grade <= 100:
                return grade
            else:
                print("Grade must be between 0 and 100.")
        except ValueError:
            print("Invalid input. Please enter a number.")

# Menu
def menu():
    while True:
        print("\n Student Grade Management System")
        print("1. Add Student")
        print("2. Update Student")
        print("3. Delete Student")
        print("4. View Students")
        print("5. Exit")

        choice = input("Enter your choice: ").strip()

        if choice == "1":
            name = input("Enter the student name: ").strip()
            grade = get_grade_input()
            add_student(name, grade)

        elif choice == "2":
            name = input("Enter the student name to update: ").strip()
            grade = get_grade_input()
            update_student(name, grade)

        elif choice == "3":
            name = input("Enter the student name to delete: ").strip()
            delete_student(name)

        elif choice == "4":
            display_all_students()

        elif choice == "5":
            print("Closing program...")
            break

        else:
            print("Invalid option, please try again.")

menu()
```
